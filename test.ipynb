```python
import numpy as np
import matplotlib.pyplot as plt
from utils import cpd
```

```python
# test signal
test_signal = np.column_stack([
    np.concatenate([np.random.normal(m, 0.3, 5) for m in [0, 2, 4]]),
    np.concatenate([np.random.normal(m, 0.3, 5) for m in [5, 0, 3]])
]) % (2 * np.pi)

# plot signal
n_dim = len(test_signal[0])
fig, axs = plt.subplots(1, n_dim, figsize=(12, 2), sharex=True, sharey=True)
for i in range(n_dim):
    axs[i].scatter(np.arange(len(test_signal)), test_signal[:, i], color='black', label='signal')
    axs[i].set_xlabel("time step")
    axs[i].set_title(f"dimention {i+1}")
    axs[i].legend(loc='best')
axs[0].set_ylabel("angle (radian)")
plt.tight_layout()
plt.show()
```

run the cpd: 5 options for `distance_kind`: `euclid`, `circular`, `sphere`, `grassmann`, `spd_affine`

```python
chpnts = cpd(signal = test_signal,
             distance_kind = 'circular', 
             n_states = 5, 
             pen = 1.0)

print(chpnts)
```

```python
chpnts = cpd(signal = test_signal,
             distance_kind = 'euclid', 
             n_states = 5, 
             pen = 2.0)

print(chpnts)
```
